```python
import yfinance as yf
import datetime
import pandas as pd
import matplotlib
import numpy as np
import statsmodels as sm
```

```python
assets= pd.read_html("https://finance.yahoo.com/crypto/?offset=0&count=100")[0]['Symbol'].values
data = []
asset_names = []
```

```python
assets
```

```python
for asset in assets:
    data.append(yf.download(asset, period = "8y"))
```

```python
data
```

```python
volume= pd.concat(data, axis = 1)['Volume']
volume.columns = assets
```

```python
close_price = pd.concat(data, axis = 1)['Close']
close_price.columns = assets
```

```python
close_price
```

```python
returns = close_price.copy()
for i in np.arange(0,len(close_price)):
    if i == 0:
        returns.iloc[i]= 0
    else:
        returns.iloc[i]= np.log(close_price.iloc[i]/close_price.iloc[i-1])
# returns.drop("2015-08-10", inplace=True)
```

```python
returns['BTC-USD'].ewm(span=len(returns['BTC-USD'])).std().plot()
```
