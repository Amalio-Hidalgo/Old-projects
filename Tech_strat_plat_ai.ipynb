```python
# How do customers’ WTP for AV-based rides vary by their gender or age group?
func = lambda x: 'Under 30' if x < 30 else ('30-65' if x < 65 else 'Over 65')
raw_data['Age_Group'] = raw_data['age'].apply(func)
raw_data.groupby(['Age_Group','Sex'])['maxfare_2'].describe().sort_values(by='mean', ascending=False)[['mean','std']]
```

```python
# Which companies do consumers prefer to be the leaders in the AV industry? 
# pd.DataFrame(raw_data[['company_purchase_tesla', 'company_purchase_gm','company_purchase_toyota', 
#                        'company_purchase_google','company_purchase_apple', 'company_purchase_uber','company_purchase_lyft']].mean().sort_values(ascending=True), columns=['average_ranking'])

# How do consumers’ preferences change if they are very tech savvy vs. not very tech savvy?
raw_data['tech_savvy'] = raw_data['firstmover_1'].apply(lambda x: 'Not Tech Savvy' if (x == 'Disagree' or x == 'Strongly Disagree') else ('Tech Savvy' if (x == 'Agree' or x == 'Strongly Agree') else 'Neutral'))
raw_data.groupby('tech_savvy')[['company_purchase_tesla', 'company_purchase_gm','company_purchase_toyota', 'company_purchase_google','company_purchase_apple', 'company_purchase_uber','company_purchase_lyft']].mean().transpose().sort_values(by='Tech Savvy')
```
