```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Data for Creators
creator_attributes = [
    'Direct IP Control',
    'Legal Precedent',
    'Potential Addressable Market',
    'IP Protection Strength',
    'Revenue Distribution',
    'Accesibility',
    'Transparency',
    'Ease of Use',
    'Cost Efficiency',
    'IP Composability & Interoperability',
    'Institutional Trust',

]

story_protocol_creators = pd.Series([9, 2, 9, 8, 9, 10, 10, 7, 9, 9, 2], creator_attributes).sort_index()
traditional_creators = pd.Series([6, 9, 6, 9, 6, 3, 5, 4, 3, 4, 8], creator_attributes).sort_index()

# Setup Creator Plot
plt.figure(figsize=(15, 12))
plt.subplot(2, 1, 1)
plt.plot(story_protocol_creators, 'b-o', linewidth=2, label='Story Protocol', color='#2E86C1')
plt.plot(traditional_creators, 'r-o', linewidth=2, label='Traditional IP Systems', color='#E74C3C')

plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Value Attributes', fontsize=10, fontweight='bold')
plt.ylabel('Performance Score (1-10)', fontsize=10, fontweight='bold')
plt.title('Creator Value Curve', fontsize=12, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper right', frameon=True, shadow=True)

# Data for Users
user_attributes = [
    'IP Discovery Speed',
    'IP Composability & Interoperability',
    'Cost Efficiency',
    'Ease of Use',
    'Legal Precedent',
    'Institutional Trust',
    'IP Protection Strength',
    'AI Integration',
    'Revenue Distribution',
    'Accesibility',
    'Transparency',

]

story_protocol_users = pd.Series([9, 9, 9, 7, 2, 2, 8, 9, 9, 10, 10], user_attributes).sort_index()
traditional_users = pd.Series([6, 5, 3, 4, 10, 8, 9, 4, 3, 5, 7], user_attributes).sort_index()


# Setup User Plot
plt.subplot(2, 1, 2)
plt.plot(story_protocol_users, 'b-o', linewidth=2, label='Story Protocol', color='#2E86C1')
plt.plot(traditional_users, 'r-o', linewidth=2, label='Traditional IP Systems', color='#E74C3C')

plt.grid(True, linestyle='--', alpha=0.7)
plt.xlabel('Value Attributes', fontsize=10, fontweight='bold')
plt.ylabel('Performance Score (1-10)', fontsize=10, fontweight='bold')
plt.title('User Value Curve', fontsize=12, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper right', frameon=True, shadow=True)

plt.tight_layout()
plt.show()
```
